```python
%matplotlib inline
import numpy as np 
import matplotlib.pyplot as plt
```

# Solving for bound, unbound and critical universes using parametric equations and numerical integration 

##  We learned that we can parametrize the equations that tell us how the universe will evolve using eta 
First I defined Eta. Then I plugged it into to the unbound parametrized equation (Omega<1) , and then the bound universe's parametrized equation (Omega>1)
This way I had values for tau and xi.
For the critical case (omega=1) I just used the equation that we derived in class. 
Finally I plotted these values of tau and xi to see how the universe will evolve in diferrent situations 

```python
eta = np.linspace(0,2*np.pi,100)
print(eta[0])
#print (eta)
A_bound=(1/2.)*(1-np.cos(eta))
T_bound=(1/2.)*((eta)-np.sin(eta))
plt.plot(T_bound,A_bound, label= "Bound Universe")
```

## This is the how the universe will evolve for the bound case. I defined eta from 0 to 2pi, then I found my xi and tau values and I plotted them.

```python
A_unbound=(1/2.)*(np.cosh(eta)-1)
T_unbound=(1/2.)*(np.sinh(eta)-eta)
plt.plot(T_unbound,A_unbound, label= "Unbound Universe")
```

# This is the case for the unbound universe. I used the same eta value, ranging from 0 to 2pi but this time I used equations for the unbound case to find my xi and tau values. 

```python
A_critical = (3/2)*(eta)**(2/3.)
plt.plot(eta, A_critical, label= "Critical Universe")
```

## This is the critical case, which omega =1. In this case we know that xi=3/2(eta)^2/3 so we just plot xi vs eta and find how the universe would evolve in a critical condition.  

```python
plt.plot(eta, A_critical, label= "Critical Universe") 
plt.plot(T_unbound,A_unbound, label= "Unbound Universe")
plt.plot(T_bound,A_bound, label= "Bound Universe")
plt.xlim(0,3)
plt.ylim(0,1)
plt.legend (loc="lower center")
plt.xlabel("Time")
plt.ylabel("Size of Universe")
plt.title ("3 possible Fates of the Universe")
```

## This graph show us how the three fates of the universe compare, and how the size of the universe would have to evolve for each case. 

```python
import scipy
from scipy.integrate import quad 
def integrand_unbound(x):
    return (x/(1+x))**(1/2.)
def integrand_bound(x):
    return (x/(1-x))**(1/2.)
xi=np.linspace(0,1,100)
t_unbound = np.zeros(len(xi))
t_bound = np.zeros(len(xi))
for i in range(len(xi)):
    t_unbound[i]=scipy.integrate.quad(integrand_unbound, 0, xi[i])[0]
    t_bound[i]=scipy.integrate.quad(integrand_bound, 0, xi[i])[0]
#print (t_unbound)
#print(t_bound)
plt.plot(t_unbound, xi,"bo")
plt.plot(t_bound,xi,"k+")
plt.plot(eta, A_critical, label= "Critical Universe") 
plt.plot(T_unbound,A_unbound, label= "Unbound Universe")
plt.plot(T_bound,A_bound, label= "Bound Universe")
plt.xlim(0,3)
plt.ylim(0,1) 
plt.legend (loc="lower center")
```

## This graph shows us how the numerical integration values compare to the parametrized equation values. We can see that they are a perfect match. So the numerical integration method also works to solve for the bound and unbound cases.
